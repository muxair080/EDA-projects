```python
"""
# If we have some duration time less than hour and given
in mints so first we can convert it to hrs by using this
code.
"""
hrs = []
for i in range(len(final_df)):
    val =  final_df.at[i,'Duration'].split(' ')[0]
    for j in range(0,59):
        if val == str(j)+'m':            
            mints = val.split('m')
            mints = int(mints[0])
            hr = "%.2f" %(mints/60)
            hrs.append(hr)
            final_df.at[i,'Duration'] =  str(hr)+'h'
```

```python
duration = final_df['Duration'].str.split(' ')
duration_hr = duration.str[0]
duration_hr = duration_hr.str[0]
duration_hr = duration_hr.astype(int)
duration_hr_to_mints =  duration_hr* 60;
duration_mints =  duration_hr_to_mints  
duration_mints = duration.str[1].str[0]
duration_mints
```

# Convert Categorical Features of airline column to Numerical values
